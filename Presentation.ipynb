```python
import pandas as pd

# Sample dataset
data = {
    'Month': ['Shrawan', 'Bhadra', 'Ashwin', 'Kartik', 'Mangsir', 'Poush', 'Magh', 'Falgun', 'Chaitra', 'Baisakh', 'Jestha', 'Ashadh'],
    'Headline_Inflation': [8.64, 8.45, 8.50, 8.30, 8.10, 7.95, 7.80, 7.65, 7.50, 7.10, 6.85, 6.83],
    'Food_Inflation': [6.5, 6.4, 6.3, 6.8, 6.9, 7.0, 6.7, 6.2, 5.9, 5.2, 4.8, 4.5],
    'NonFood_Inflation': [9.2, 9.1, 9.4, 9.3, 9.0, 8.8, 8.7, 8.6, 8.3, 8.1, 8.0, 8.0]
}
df = pd.DataFrame(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Reshape data for seaborn
df_melted = df.melt(id_vars='Month', value_vars=['Food_Inflation', 'NonFood_Inflation'],
                    var_name='Category', value_name='Inflation')

plt.figure(figsize=(8, 6))
sns.boxplot(data=df_melted, x='Category', y='Inflation', palette='Set2')
plt.title('Box Plot of Food vs Non-Food Inflation')
plt.ylabel('Inflation Rate (%)')
plt.xlabel('Category')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Headline_Inflation'], label='Headline Inflation', marker='o')
plt.plot(df['Month'], df['Food_Inflation'], label='Food Inflation', marker='s')
plt.plot(df['Month'], df['NonFood_Inflation'], label='Non-Food Inflation', marker='^')
plt.title('Monthly Inflation Trends (2022–23)')
plt.xlabel('Month')
plt.ylabel('Inflation Rate (%)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x=df['Food_Inflation'], y=df['NonFood_Inflation'], hue=df['Month'], palette='viridis', s=100)
plt.title('Scatter Plot: Food vs Non-Food Inflation')
plt.xlabel('Food Inflation (%)')
plt.ylabel('Non-Food Inflation (%)')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
